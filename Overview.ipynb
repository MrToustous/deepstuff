```python
def get_songs(genre: int = 4):
    mask = tracks.genres.map(lambda x: genre in x, na_action='ignore')
    return tracks[mask.replace(np.NaN, False)].copy()
```

# Extract 

```python
import zipfile
```

```python
def id_path(song_id: int = None):
    id_str = str(song_id)
    song_number = id_str
    for i in range(6-len(song_number)):
        song_number = '0' + song_number
    file_number = id_str[:-3]
    for i in range(3-len(file_number)):
        file_number = '0' + file_number
    return file_number + '/' + song_number + '.mp3'
```

```python
songs = techno.index.get_level_values(0)
```

```python
with zipfile.ZipFile(DATA_PATH + 'fma_large.zip', 'r') as zip_ref:
    for song in songs:
        zip_ref.extract('fma_large/'+id_path(song), 'songs/electro')
#     temp = zip_ref.namelist()
```
